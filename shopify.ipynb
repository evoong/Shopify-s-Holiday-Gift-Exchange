```python
import random

class Hat:
    def __init__(self):
        self._particpants = []

    def add_name(self, particpant):
        self._particpants.append(particpant)

    def shuffle(self):
        random.shuffle(self._particpants)

    def draw(self):
        return self._particpants.pop()
    
    def get_participants(self):
        return self._particpants
    
class Participants():
    def __init__(self, name, partner):
        self.name = name
        self.partner = partner

    def get_name(self):
        return self.name
    
    def get_partner(self):
        return self.partner


class Game():
    def __init__(self):
        self._hat = Hat()
        self._participants = []
        self._past_draws = []
    

    def add_participant(self, name, partner):
        particpant = Participants(name, partner)
        self._participants.append(particpant)
        self._hat.add_name(particpant)



    def draw(self, particpant):
        self._hat.shuffle()
        draw = self._hat.draw()

        # grab your own name
        if draw.get_name() == particpant.get_name() :
            self._hat.add_name(draw)
            return False
        
        #grab your partners name
        if draw.get_partner() == particpant.get_name() and particpant.get_partner() != "":
            self._hat.add_name(draw)
            return False
        
        #grab a past draw
        if draw.get_name() in [x[0].get_name() for x in self._past_draws]:
            return False
        
        # grab a past draw
        if draw.get_partner() in [x[1].get_name() for x in self._past_draws]:
            return False

        
        else:
            # store the draw
            self._past_draws.append([particpant,draw])

            # remove the particpant from the list of participants
            for p in self._participants:
                if draw.get_name() == p.get_name():
                    self._participants.remove(p)
            return True


    def get_hat(self):
        return self._hat
    
    def game_over(self):
        return len(self._participants) <= 1
    
    def get_participants(self):
        return self._participants  
    
    def get_past_draws(self):    
        return self._past_draws
    
    def print_past_draws(self):
        for x in self._past_draws:
            print(x[0].get_name() + ' drew ' + x[1].get_name())

class game_demo():
    def __init__(self):
        self._game = Game()


    def run(self):
        self._game.add_participant('John', 'Jane')
        self._game.add_participant('Kevin', 'Bob')
        self._game.add_participant('Steve', 'Eric')
        self._game.add_participant('Jane', 'John')

        
        
        while True:
            current_player = self._game._participants.pop()
            print(current_player.get_name() + ' is drawing')
            while True:
                if self._game.draw(current_player):
                    print(self._game.get_past_draws()[-1][0].get_name() + ' drew ' + self._game.get_past_draws()[-1][1].get_name())
                    break
            if self._game.game_over():
                break
        print('Game Over!') 
        print("Final Assignment")
        self._game.print_past_draws() 



    

if __name__ == '__main__':
    game_demo().run()


        
```
